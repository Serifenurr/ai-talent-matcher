```python
import os
import docx


klasordeki_dosyalar = os.listdir('.')
docx_dosyalari = [dosya for dosya in klasordeki_dosyalar if dosya.endswith('.docx')]
print(f"Toplam {len(docx_dosyalari)} adet word formatında cv bulundu!\n")

test_cv_listesi = []
cv_isimleri = []

for dosya_adi in docx_dosyalari[:5]:
    try:
        doc = docx.Document(dosya_adi)
        tam_metin = []
        for paragraf in doc.paragraphs:
           
            if paragraf.text.strip() != "":
                tam_metin.append(paragraf.text)
        
      
        if tam_metin:
            birlestirilmis_metin = '\n'.join(tam_metin)
            test_cv_listesi.append(birlestirilmis_metin)
            cv_isimleri.append(dosya_adi)
            
    except Exception as e:
        print(f"{dosya_adi} okunurken hata oluştu: {e}")


if len(cv_isimleri) > 0:
    print("Dosyalar başarıyla metne çevrildi!")
    print(f"--- Örnek: {cv_isimleri[0]} dosyasının başlangıcı ---")
    print(test_cv_listesi[0][:200] + "...\n")
else:
    print("HATA: Hiçbir dosya okunamadı. Lütfen klasörde geçerli .docx dosyaları olduğundan emin ol.")
```

```python
from sentence_transformers import SentenceTransformer

model= SentenceTransformer('all-MiniLM-L6-v2')
print("Yapay Zeka Modeli(embedding Engine) başarıyla yüklendi")
```

```python
ceo_kriterleri="AA software engineer with experience in React,Node.js,and MongoDB.Familiar with full-stack development."
ceo_vektoru=model.encode([ceo_kriterleri])
print(f"CEO'nun beklentisi {ceo_vektoru.shape[1]} boyutlu bir sayı dizisine dönüştü.")
```

```python
cv_vektorleri=model.encode(test_cv_listesi)
print(f"{len(test_cv_listesi)} adet CV başarıyla vektöre dönüştürüldü.")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
benzerlik_skorlari=cosine_similarity(ceo_vektoru,cv_vektorleri)[0]

eslesme_listesi=[]
for i in range(len(cv_isimleri)):
    eslesme_listesi.append(
        {
            "Dosya":cv_isimleri[i],
            "Skor":round(benzerlik_skorlari[i]*100,2)
        }
    )
    sirali_liste=sorted(eslesme_listesi,key=lambda x:x['Skor'],reverse=True)

print("---AI TALENT MATCHER SONUCLARI---\n")
for aday in sirali_liste:
    print(f"Aday Dosyası:{aday['Dosya']} | Uyumluluk Skoru:%{aday['Skor']}")
```
